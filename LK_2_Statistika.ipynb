```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from numpy import *
from statistics import *
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/arawsardni/Statistika/main/biodata_tif_d.csv")

df
```

```python
# Preprocessing
df["Tinggi Badan"] = df["Tinggi Badan"].str.replace("cm", "").astype(int)
df["Umur"] = df['Umur'].str.replace('Tahun', '').str.replace('tahun', '').dropna().astype(int)
df['Asal'] = df['Asal'].str.strip()
df['cita-cita'] = df['cita-cita'].str.strip()
```

```python
# 5 Mahasiswa tertinggi di kelas Statistika TIF-D 2024
df.sort_values(["Tinggi Badan"], ascending = False).head(5)
```

```python
# 5 Mahasiswa terpendek di kelas Statistika TIF-D 2024
df.sort_values(["Tinggi Badan"], ascending = False).tail(5)
```

```python
# Data Describe
tinggi_badan = df["Tinggi Badan"]
min_tinggi = min(tinggi_badan)
max_tinggi = max(tinggi_badan)
mode_tinggi = mode(tinggi_badan)
mean_tinggi = mean(tinggi_badan)
median_tinggi = median(tinggi_badan)

print(f"Tinggi badan minimum : {min_tinggi} cm")
print(f"Tinggi badan maksimum : {max_tinggi} cm")
print(f"Modus tinggi badan : {mode_tinggi} cm")
print(f"Rata-rata tinggi badan : {mean_tinggi} cm")
print(f"Median tinggi badan : {median_tinggi} cm")
```

```python
# Visualisasi Tinggi Badan
x = df['Tinggi Badan']
plt.hist(x, bins = 8, rwidth=0.9)
plt.title('Distribusi Tinggi Badan')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Frekuensi')
plt.hist
```

```python
# Asal Kota
plt.figure(figsize=(8, 6))
plt.pie(df["Asal"].value_counts(), labels = df["Asal"].value_counts().index, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart Asal Kota")
plt.axis("equal")
plt.show()
```

```python
# Grafik distribusi umur
plt.figure(figsize=(10, 6))
sns.histplot(df["Umur"], bins=10, kde=True)
plt.title('Distribusi Umur')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
# Grafik distribusi cita-cita
plt.figure(figsize=(12, 6))
sns.countplot(y='cita-cita', data=df, order=df['cita-cita'].value_counts().index)
plt.title('Distribusi Cita-cita')
plt.xlabel('Frekuensi')
plt.ylabel('Cita-cita')
plt.show()
```

```python
# Konversi kolom tanggal lahir menjadi tahun lahir
df['Tahun Lahir'] = pd.to_datetime(df['tanggal lahir'], errors='coerce').dt.year

# Hapus baris yang memiliki tahun lahir yang tidak valid
df = df.dropna(subset=['Tahun Lahir'])

# Grafik distribusi tahun lahir
plt.figure(figsize=(10, 6))
sns.histplot(df['Tahun Lahir'], bins=10, kde=True)
plt.title('Distribusi Tahun Lahir')
plt.xlabel('Tahun Lahir')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Asal', y='Tinggi Badan', palette='viridis')
plt.title('Hubungan antara Tinggi Badan dan Asal')
plt.xlabel('Asal')
plt.ylabel('Tinggi Badan (cm)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Umur', y='Tinggi Badan', palette='viridis')
plt.title('Hubungan antara Tinggi Badan dan Umur')
plt.xlabel('Umur')
plt.ylabel('Tinggi Badan (cm)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='cita-cita', y='Tinggi Badan', palette='viridis')
plt.title('Hubungan antara Tinggi Badan dan Cita-Cita')
plt.xlabel('Cita-CIta')
plt.ylabel('Tinggi Badan (cm)')
plt.xticks(rotation=45)
plt.show()
```
